```python
# Determine a distância entre as retas paralelas r e s nos seguintes casos

# Se as retas são paralelas, obrigatoriamente, os coeficientes a e b serão iguais.
import math
print('Digite os termos da equação, em caso de inexistÊncia, insira o valor 0')
print('''
O programa só recebe números no formato float (decimal), é necessário
que em caso de frações os números estejam em sua forma decimal.
Para converter esses números utilize a calculadora no bloco seguinte

''')
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
cr = float(input("Digite o valor de c da reta r: "))
cs = float(input("Digite o valor de c da reta s: "))

distancia = abs(cr - cs)/math.sqrt(a**2 + b**2)


distanciaFormatada = "{:.2f}".format(round(distancia, 2))

#RESULTADO
print('''A distância entre as retas paralelas
é: {}'''.format(distanciaFormatada))
```

```python
#CALCULADORA 
a, b = input('Digite separado por vírgula os números que você deseja dividir, exemplo (a,b)').split(",")

a, b = int(a), int(b)

solucao = a/b
solucaoformat = "{:.2f}".format(round(solucao, 2))

print('O resultado da divisão {}/{} é {}'.format(a, b, solucaoformat))
```
